# üìù To-Do List Task Manager (Interactive Google Colab Notebook)
This notebook provides a simple console-based task manager written in Python.
- Add, list, edit, and delete tasks
- Tasks are saved to a `.txt` file (`tasks.txt`)
- You can download your saved task list after you're done

**Run each cell in order and interact with the menu using inputs.**

```

# Task Manager Application (To-Do List)
# This app allows users to add, view, edit, and delete tasks using a text file

# Load tasks from file
def load_tasks(filename="tasks.txt"):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            return [line.strip() for line in file.readlines()]
    except FileNotFoundError:
        return []  # Return an empty list if file doesn't exist
    except Exception as e:
        print("Error while loading tasks:", e)
        return []

# Save tasks to file
def save_tasks(tasks, filename="tasks.txt"):
    try:
        with open(filename, "w", encoding="utf-8") as file:
            for task in tasks:
                file.write(task + "\n")
    except Exception as e:
        print("Error while saving tasks:", e)

# Add a new task
def add_task(tasks):
    task = input("Enter a new task: ").strip()
    if task == "":
        print("‚ö†Ô∏è Cannot add an empty task.")
    else:
        tasks.append(task)
        print("‚úÖ Task added successfully.")

# List all tasks
def list_tasks(tasks):
    if not tasks:
        print("üìÇ No tasks found.")
    else:
        print("\nüìù Task List:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
        print()

# Delete a task by number
def delete_task(tasks):
    list_tasks(tasks)
    try:
        choice = int(input("Enter the number of the task to delete: "))
        if 1 <= choice <= len(tasks):
            removed = tasks.pop(choice - 1)
            print(f"üóëÔ∏è '{removed}' has been deleted.")
        else:
            print("‚ö†Ô∏è Invalid task number.")
    except ValueError:
        print("‚ö†Ô∏è Please enter a valid number.")

# Edit a task by number
def edit_task(tasks):
    list_tasks(tasks)
    try:
        choice = int(input("Enter the number of the task to edit: "))
        if 1 <= choice <= len(tasks):
            new_task = input("Enter the new task text: ").strip()
            if new_task == "":
                print("‚ö†Ô∏è Cannot enter an empty task.")
            else:
                old = tasks[choice - 1]
                tasks[choice - 1] = new_task
                print(f"‚úèÔ∏è '{old}' updated to '{new_task}'.")
        else:
            print("‚ö†Ô∏è Invalid task number.")
    except ValueError:
        print("‚ö†Ô∏è Please enter a valid number.")

# Main menu function
def main_menu():
    tasks = load_tasks()

    while True:
        print("\nüîπ Task Manager Menu üîπ")
        print("1. List Tasks")
        print("2. Add New Task")
        print("3. Edit Task")
        print("4. Delete Task")
        print("5. Exit")

        choice = input("Choose an option (1-5): ").strip()

        if choice == "1":
            list_tasks(tasks)
        elif choice == "2":
            add_task(tasks)
            save_tasks(tasks)
        elif choice == "3":
            edit_task(tasks)
            save_tasks(tasks)
        elif choice == "4":
            delete_task(tasks)
            save_tasks(tasks)
        elif choice == "5":
            print("üìÅ Saving and exiting the application...")
            save_tasks(tasks)
            break
        else:
            print("‚ö†Ô∏è Invalid choice! Please enter a number between 1 and 5.")
```

```
main_menu()
```
